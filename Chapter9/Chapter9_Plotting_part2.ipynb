```python
from datetime import datetime
```

```python
fig = plt.figure()
```

```python
ax = fig.add_subplot(1,1,1)
```

```python
data = np.random.randn(1000).cumsum()
```

```python
ax.plot(data,'k-')
```

```python
crisis_data =[(250.,'Peak of internet market'),
              (470.,'Constant of life'),
              (800., 'Chab en Bourse')
             ]
```

```python
for date,label in crisis_data:
    ax.annotate(label, xy=(date,5), xytext=(date,10), arrowprops=dict(facecolor='black', headwidth=4, width=2, headlength=4),
               horizontalalignment='left',verticalalignment='top')
```
